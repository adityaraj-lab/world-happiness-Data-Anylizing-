# there is no missing value in our dataset in this case our data is robust 

```python
df.head()
new_df1=df[['Country name','Ladder score']]
new_df2=new_df1.sort_values('Ladder score',ascending=False).head(20)
```

# check top_20 country which have high Ladder Score

Ladder score - Life evaluation score

```python
fig,ax=plt.subplots(1,1,figsize=(10,5))
sns.barplot(x='Ladder score',y='Country name',data=new_df2)
```

```python
new_df3=new_df1.sort_values('Ladder score',ascending=True).head(20)
plt.figure(figsize=(10,5))
sns.barplot(x='Ladder score',y='Country name',data=new_df3)
```

# Country vs GDP per capita

Logged.GDP.per.capita - Extent to which GDP contributes to the calculation of the Ladder score

```python
new_df4=df[['Country name','Logged GDP per capita']].sort_values('Logged GDP per capita',ascending=False).head(20)
plt.figure(figsize=(10,5))
sns.barplot(x='Logged GDP per capita',y='Country name',data=new_df4)
```
